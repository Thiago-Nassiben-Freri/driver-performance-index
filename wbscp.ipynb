```python
import requests 
from bs4 import BeautifulSoup


url = "https://www.statsf1.com/pt/statistiques/pilote/point/nombre.aspx"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find_all("table")
target_table = table[2]

print(target_table)
```

```python
head = []
rows = []

for td in target_table.find("thead").find_all("td"):
    head.append(td.get_text(strip=True))

for td in target_table.find("tbody").find_all("td"):
    rows.append(td.get_text(strip=True))

dict_data = {chave: [] for chave in head}

for i in range(0, len(rows), len(head)):
    bloco = rows[i:i+len(head)]
    for chave, valor in zip(head, bloco):
        dict_data[chave].append(valor)

print(dict_data)
```

```python
import pandas as pd


df = pd.DataFrame(dict_data)

df
```

```python
df.dtypes
```

```python
df
```

```python
df["Moy/GP"] = df["Moy/GP"].astype(float)

df
```

```python
df.dtypes
```

```python
df["Nb"] = [float(p.replace(" ", "")) for p in df["Nb"]]

df.dtypes
```
